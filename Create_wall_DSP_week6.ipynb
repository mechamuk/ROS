# Empty world

* 벽이 없는 gazebo 실행
    * roslaunch turtlebot3_gazebo turtlebot3_empty_world.launch

* src 폴더에 패키지 폴더를 생성
* 나의 경우 패키지 폴더 이름을 dsp로 생성
    * cd catkin_ws/src
    * catkin_create_pkg dsp

* 패키지 폴더에 런치파일과 world 파일을 만들 두 폴더 생성
    * 각각 launch와 worlds로 생성

* 각 폴더 내에서 empty document 파일 생성
    * 나의 경우 new_map.launch와 new_world.world 생성 

* turtlebot3_empty_world.launch을 new_map.launch에 복사
* <arg name="world_name" value=" 의 뒷부분을 수정
    * (find turtlebot3_gazebo)/worlds/empty.world"/>
    * (find dsp)/worlds/new_world.world"/>

* empty.world를 new_world.world에 복사

* workspace에서 빌드
    * cd catkin_ws
    * catkin_make

* 실행
    * roslaunch dsp new_map.launch

* 잘 안된다면 잘 수정했는지 확인하고
* workspace나 소스 코드로 이동해서 다음을 진행 후 다시 실행
    * rospack list

# 맵 만들기

* 왼쪽 상단 탭에서 edit -> building editor를 실행
* 원하는대로 벽을 세우고 저장(save as)
* model name 설정 후
    * location을 home/(자신의 컴퓨터 이름)/.gazebo/models/(model name)
    * ex) model name = test
    * ex) /home/muk/.gazebo/models/test

* 새로운 프로젝트를 열 때마다 실행시키기 위해서는
* new_world.world 파일 수정
* A ground plane 부분 4줄 복붙 후 다음과 같이 수정
    * A ground plane -> test
    * model://ground_plane -> model://test

* 다시 빌드
    * cd catkin_ws
    * catkin_make

* 다시 실행하여 벽이 생기는지 확인
    * roslaunch dsp new_map.launch
